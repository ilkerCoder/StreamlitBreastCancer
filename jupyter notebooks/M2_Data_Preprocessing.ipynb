```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
sns.displot(df["radius_mean"])
```

```python
"""
DATA ANALYZE KISMINDA FARKINA VARDIGIMIZ BİR NaN ' LARLA DOLU COLUMN VARDI.
EKSTRA OLARAK GEREKSİZ COLUMNLARI DA SİLMEMİZ GEREK. İLK ONCE O İŞLEMLERLE
BAŞLAYALIM.
"""
dftemp=df.drop("id" , axis=1)
dftemp
```

```python
print("SÜTUN SAYISI ==>" ,  len(dftemp.columns))
""" 32 SÜTUNDAN SADECE 1 TANESİNDE NULL DEGERLER VAR . SİMDİ BUNA NE
YAPACAGIMIZA KARAR VERECEGİZ. """
pd.DataFrame(dftemp.isnull().sum())
```

```python
"""SONDAKİ TAMAMEN NAN OLAN SÜTUNU DA KALDIRDIK . """
dftemp.drop("Unnamed: 32" , inplace=True , axis=1)
pd.DataFrame(dftemp.columns)
```

```python
""" ARTIK NULL DEGERLERIMIZ YOK ."""
dftemp.isnull().sum()
```

```python
"""Gercekten labelimiz daki tüm deger m ya da b mi ? """
diagnosis_column = dftemp["diagnosis"]
diagnosis_column.value_counts()
```

```python
"""DİAGNOSİS SÜTUNUNDAKİ M DEGERİNİ 1 B DEGERİNİ 0 OLACAK SEKİLDE
    GÜNCELLE"""
dftemp['diagnosis'] = dftemp['diagnosis'].replace({'M': 1, 'B': 0})
dftemp["diagnosis"].value_counts() #Basarılı
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(dftemp.corr())
```

```python
dftemp.corr()["diagnosis"][:-1].sort_values().plot(kind="bar")
```

```python
""" 
GORULDUGU UZERE BIRKAC DEGERIMIZ 0 AMA BU DEGERLERİN DİREK 0 OLMASI
PEK OLASI DEGİL. BU DEGERLER İCİN BIRKAC COZUM YOLU VAR.
Fakat Once kac adet 0 degerimiz var columnlar icin bir bakalım
"""
pd.DataFrame(dftemp.tail(10))
```

```python
"""
her columndaki her satır icin kac tane 0 degeri var 
diagnosis haric nasıl bir bakalım
""" 
zero_counts = dftemp.isin([0]).sum()
pd.DataFrame(zero_counts)

```

```python
numeric_columns = dftemp.drop(columns=['diagnosis'])

# Sıfır değerlerini ortalama ile doldurma
for column in numeric_columns.columns:
    if (dftemp[column] == 0).any():  # Sıfır değeri varsa
        mean_value = dftemp[column].replace(0, pd.NA).mean()  # 0 değerleri geçici olarak NaN ile değiştir ve ortalama al
        dftemp[column] = dftemp[column].replace(0, mean_value)
```

```python
#0 OLAN DEGERLERİ ORTALAMAYLA DOLDURDUK .
zero_counts = dftemp.isin([0]).sum()
pd.DataFrame(zero_counts)
```

```python
data = dftemp[['radius_mean', 'texture_mean', 'diagnosis']]

malignant_data = data[data['diagnosis'] == 0]
benign_data = data[data['diagnosis'] == 1]

plt.figure(figsize=(10, 6))
sns.set_theme(style="dark")
sns.scatterplot(data=malignant_data, x='radius_mean', y='texture_mean', color='red', label='KOTU',alpha=0.3 )
sns.scatterplot(data=benign_data, x='radius_mean', y='texture_mean', color='blue', label='IYI', alpha=0.3 )

plt.title('Scatter Plot of Radius Mean vs Texture Mean')
plt.xlabel(xlabel='Radius Mean')
plt.ylabel('Texture Mean')
plt.legend()
plt.show()
```
